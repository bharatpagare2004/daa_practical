```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error

# Load data
df = pd.read_csv("uber.csv")

# Keep only needed columns and drop missing rows
df = df[['fare_amount','pickup_longitude','pickup_latitude',
         'dropoff_longitude','dropoff_latitude','passenger_count']].dropna()

# Simple distance formula
df['distance'] = ((df['dropoff_latitude']-df['pickup_latitude'])**2 + 
                  (df['dropoff_longitude']-df['pickup_longitude'])**2) ** 0.5

# Input and output
X = df[['distance','passenger_count']]
y = df['fare_amount']

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Train models
lr = LinearRegression().fit(X_train, y_train)
rf = RandomForestRegressor().fit(X_train, y_train)

# Predict
p1, p2 = lr.predict(X_test), rf.predict(X_test)

# Evaluate
print("Linear R2:", r2_score(y_test, p1))
print("RF R2:", r2_score(y_test, p2))
print("Linear RMSE:", mean_squared_error(y_test, p1, squared=False))
print("RF RMSE:", mean_squared_error(y_test, p2, squared=False))
```
