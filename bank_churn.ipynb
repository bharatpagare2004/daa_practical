```python
#CreditScore Age Tenure Balance NumOfProducts HasCrCard IsActiveMember EstimatedSalary
x = data[['CreditScore','Age','Tenure','Balance','NumOfProducts','HasCrCard','IsActiveMember','EstimatedSalary']]
```

```python
y = data["Exited"]
```

```python
#normalization using standard scalar
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
scaler =StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,random_state = 0 , test_size = 0.25)
```

```python
x.shape
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
# for neural_network
from sklearn.neural_network import MLPClassifier
```

```python
ann = MLPClassifier(hidden_layer_sizes = (100,100,100),activation = 'relu',random_state = 0,max_iter = 100)
ann.fit(x_train,y_train)
y_pred_ann = ann.predict(x_test)
```

```python
y_test.value_counts()
```

```python
from sklearn.metrics import accuracy_score , classification_report , confusion_matrix
```

```python
print("accuracy:",accuracy_score(y_test , y_pred_ann))
```

```python
print("confusion_matrix:",confusion_matrix(y_test, y_pred_ann))
```

```python
print("classification_report :",classification_report(y_test,y_pred_ann))
```
